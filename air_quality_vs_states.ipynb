```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('3b01bcb8-0b14-4abf-b6f2-c1bfd384ba69.csv')
df = df.drop(columns=['id','pollutant_unit'])
new_df = df.groupby('state').mean()
id_df = df.groupby('pollutant_id').mean()
```

```python
states = [i for i in new_df.index]
Pollutant_ID = [i for i in id_df.index]
```

```python
plt.figure(figsize=(20,8))
sns.set_theme(style = 'darkgrid',font_scale=1.25)
plt.xticks(rotation=90)
plt.plot(new_df['pollutant_avg'],label='Pollutant Avg.',color='red',linestyle='solid',linewidth=3,marker = 'o',mfc='b')
plt.plot(new_df['pollutant_min'],label='Pollutant Min.',color='blue',linestyle='solid',linewidth=3,marker = 'o',mfc='w')
plt.plot(new_df['pollutant_max'],label='Pollutant Max.',color='black',linestyle='solid',linewidth=3,marker = 'o',mfc='w')
plt.title('Air Quality Index')
plt.xlabel('States')
plt.ylabel('Pollutants contents (micro_gm/cubic metre)')
plt.legend(loc='best')
plt.show()
```

```python
df_pie = df.groupby('pollutant_id').mean()
# df_pie
# df_pie = df[['pollutant_id','pollutant_avg']]
# df_pie.iloc[:,0]
plt.figure(figsize=(20,10))
palette_color_1 = sns.color_palette('pastel')
# # palette_color_2= sns.color_palette('dark')
explode_list=[0.1,0.2,0.1,0.2,0.1,0.2,0.1]
plt.pie(df_pie.iloc[:,2],labels=['CO',"NH3",'NO2','OZONE','PM10','PM2.5','SO2'],shadow=False,autopct='%.3f%%',colors=palette_color_1,explode = explode_list)
plt.title('Pie Chart for Different Pollutant_content % in India')
plt.show()
```
